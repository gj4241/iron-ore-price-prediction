```python
from sklearn.linear_model import Ridge
import xgboost as xgb
import time
import pandas as pd
import numpy as np
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn import cross_validation, metrics   #Additional scklearn functions
from sklearn.grid_search import GridSearchCV   #Perforing grid search
```

```python
import re
regex = re.compile(r"\[|\]|<", re.IGNORECASE)
trainX.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in trainX.columns.values]
#testX.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in testX.columns.values]
dtrain = xgb.DMatrix(trainX, trainY, feature_names=trainX.columns.values)

params = {}
params['booster']  = 'gbtree'
params['objective'] = 'reg:linear'
params['max_depth'] = 7
params['subsample'] = 0.83
params['colsample_bytree'] = 0.8
params['silent'] = 1
params['eval_metric'] = 'rmse'
num_round = 5000
eval_list  = [(dtrain,'train')]

print('training xgb model...')
bst = xgb.train(params, dtrain, num_round, eval_list)
```
